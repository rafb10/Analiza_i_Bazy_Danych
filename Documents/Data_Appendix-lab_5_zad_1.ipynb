### 2. Zmiennosc poziomu demokracji na swiecie w latach kiedy prowadzony byl ranking

```python
data['mean'] = data.mean(axis=1)
print(data['mean'])
ax = sns.barplot(data.index, data['mean'], color='green').set_title
(' Sredni poziom demokracji na calym swiecie w latach 1952-1992:')   
```

### 3. Panstwa, w ktorych poziom demokracji zostal za kazdym razem oceniony jako najwyzszy/najnizszy z mozliwych

```python
MAX_RATING = 10.0
MIN_RATING = -10.0
mostDemocraticCountries = []
mostUndemocraticCountries = []
for column in data:
    if data[column].min() ==  MAX_RATING:
        mostDemocraticCountries.append(column)
    if data[column].max() ==  MIN_RATING:
        mostUndemocraticCountries.append(column)  
        
print("Panstwa, ktorym w kazdej edycji rankingu przyznano maksymalna ocene:")
print(mostDemocraticCountries, "\n")
print("Panstwa, ktorym w kazdej edycji rankingu przyznano minimalna ocene:")
print(mostUndemocraticCountries, "\n")
```
